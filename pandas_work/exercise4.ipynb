```python
import pandas
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
d = pandas.read_html(url)[1]
d
```

```python
data = d.to_string()
data
```

```python
import re
# return the data types whose IDs are made of only 1 digit and contain at least 3 consecutive digits somewhere on their respective lines.
# professor solution: \d{1}\s{1,}(\w+?)\s{1,}.*\d{3}.*
result = re.findall(r"\n\d{1}\s+([\w\.]+)\s+.+\d{3}", data)
result
```
